```python
!ls -l /content/drive/MyDrive/research_project/data/UBFC-rPPG/DATASET_100_1_train
!ls -l /content/drive/MyDrive/research_project/data/UBFC-rPPG/DATASET_100_1_test

import os
import shutil
base_path = '/content/drive/MyDrive/research_project/data/UBFC-rPPG/DATASET_100_1_train' # @param
fold1_path = '/content/drive/MyDrive/research_project/data/UBFC-rPPG/train_fold'
fold2_path = '/content/drive/MyDrive/research_project/data/UBFC-rPPG/test_fold'

NUM_TO_COPY = 28 # @param
dirnames = os.listdir(base_path)
i = 0
for dirname in dirnames:
  if i < NUM_TO_COPY:
    if not os.path.isdir(os.path.join(fold1_path, dirname)):
      os.mkdir(os.path.join(fold1_path, dirname))
    for filename in os.listdir(os.path.join(base_path, dirname)):
      shutil.copy(os.path.join(base_path, dirname, filename), os.path.join(fold1_path, dirname, filename))
  else:
    if not os.path.isdir(os.path.join(fold2_path, dirname)):
      os.mkdir(os.path.join(fold2_path, dirname))
    for filename in os.listdir(os.path.join(base_path, dirname)):
      shutil.copy(os.path.join(base_path, dirname, filename), os.path.join(fold2_path, dirname, filename))
  i = i + 1
```

```python
base_test_path = '/content/drive/MyDrive/research_project/data/UBFC-rPPG/DATASET_100_1_test'
for dirname in os.listdir(base_test_path):
    if not os.path.isdir(os.path.join(fold2_path, dirname)):
      os.mkdir(os.path.join(fold2_path, dirname))
    for filename in os.listdir(os.path.join(base_test_path, dirname)):
      shutil.copy(os.path.join(base_test_path, dirname, filename), os.path.join(fold2_path, dirname, filename))
```
